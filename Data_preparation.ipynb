```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from PIL import Image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img=plt.imread('/content/drive/MyDrive/animal classification/images for model creation/input images/lion/lion2.jpeg')
plt.imshow(img)
```

```python
datagen=ImageDataGenerator(rotation_range=40,
                           width_shift_range=0.2,
                           height_shift_range=0.2,
                           shear_range=0.2,
                           zoom_range=0.2,
                           horizontal_flip=True,
                           fill_mode='nearest')
```

```python
for i in os.listdir('/content/drive/MyDrive/animal classification/images for model creation/input images'):  #list of foldernames kittum
  folder_path='/content/drive/MyDrive/animal classification/images for model creation/input images/' + i
  for im in os.listdir(folder_path):
    im_path=folder_path+'/'+im
    image=plt.imread(im_path)
    image=image.reshape(1,image.shape[0],image.shape[1],image.shape[2])


    if im[:3]=='cat':#first 3 letter checking
      k=1
      for batch in datagen.flow(image,batch_size=1,
                                save_to_dir='/content/drive/MyDrive/animal classification/images for model creation/augmented images',
                                save_prefix='cat',
                                save_format='jpg'):

        if k>10:
          break
        k+=1
    elif im[:3]=='dog':
      k=1
      for batch in datagen.flow(image,batch_size=1,
                                save_to_dir='/content/drive/MyDrive/animal classification/images for model creation/augmented images',
                                save_prefix='dog',
                                save_format='png'):
        if k>10:
          break
        k+=1
    elif im[:3]=='ele':
      k=1
      for batch in datagen.flow(image,batch_size=1,
                                save_to_dir='/content/drive/MyDrive/animal classification/images for model creation/augmented images',
                                save_prefix='elephant',
                                save_format='png'):
        if k>10:
          break
        k+=1
    elif im[:4]=='lion':
      k=1
      for batch in datagen.flow(image,batch_size=1,
                                save_to_dir='/content/drive/MyDrive/animal classification/images for model creation/augmented images',
                                save_prefix='lion',
                                save_format='jpg'):
        if k>10:
          break
        k+=1
    elif im[:5]=='tiger':
      k=1
      for batch in datagen.flow(image,batch_size=1,
                                save_to_dir='/content/drive/MyDrive/animal classification/images for model creation/augmented images',
                                save_prefix='tiger',
                                save_format='jpg'):
        if k>10:
          break
        k+=1
```

```python
filenamelist=[]
label=[]
for im in os.listdir('/content/drive/MyDrive/animal classification/images for model creation/augmented images'):
  image=Image.open('/content/drive/MyDrive/animal classification/images for model creation/augmented images/'+im)
  resized=image.resize((500,500))
  if im[:3]=='cat':
    file_path='/content/drive/MyDrive/animal classification/images for model creation/resized images/'+im
    filenamelist.append(file_path)
    label.append('cat')
    resized.save(file_path)
  if im[:3]=='dog':
    file_path='/content/drive/MyDrive/animal classification/images for model creation/resized images/'+im
    filenamelist.append(file_path)
    label.append('dog')
    resized.save(file_path)
  if im[:3]=='ele':
    file_path='/content/drive/MyDrive/animal classification/images for model creation/resized images/'+im
    filenamelist.append(file_path)
    label.append('elephant')
    resized.save(file_path)
  if im[:4]=='lion':
    file_path='/content/drive/MyDrive/animal classification/images for model creation/resized images/'+im
    filenamelist.append(file_path)
    label.append('lion')
    resized.save(file_path)
  if im[:5]=='tiger':
    file_path='/content/drive/MyDrive/animal classification/images for model creation/resized images/'+im
    filenamelist.append(file_path)
    label.append('tiger')
    resized.save(file_path)
```

```python
df=pd.DataFrame({'file name':filenamelist,'label':label})
```

```python
df.to_csv('/content/drive/MyDrive/animal classification/dataframe.csv',index=False)
```
